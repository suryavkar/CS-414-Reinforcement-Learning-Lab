```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rew = np.genfromtxt("a2c_cp.csv")
plt.figure(figsize=(16,10))
plt.plot(rew)
plt.title("Rewards per episodes vs Episodes")
plt.xlabel("Episodes")
plt.ylabel("Rewards per episodes")
plt.savefig("a2c_cp.jpeg")
```
