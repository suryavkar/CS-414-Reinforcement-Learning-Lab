```python
# importing packages
import numpy as np
import matplotlib.pyplot as plt

from jupyterthemes import jtplot
jtplot.style(theme='gruvboxd',context='notebook',grid=False,ticks=True)
```

## Plots

```python
# x-axis values for plotting graph
x = np.linspace(0,np.pi,1000)
```

```python
# declaring function in part a
def f1(x):
    return np.sin(x)**0.5
```

```python
# plotting function in part a
plt.figure(figsize=(10,6))
plt.plot(x,f1(x));
plt.xlabel("x");
plt.ylabel("y");
plt.title("y = sqrt(sin x)");
```

```python
# declaring function in part b
def f2(x):
    return (np.sin(x)**0.5)*(np.e**(-x**2))
```

```python
# plotting function in part b
plt.figure(figsize=(10,6))
plt.plot(x,f2(x));
plt.xlabel("x");
plt.ylabel("y");
plt.title("y = sqrt(sin x)*exp(-x^2)");
```

## Expectation calculation

```python
ns1 = 50
# sampling ns1 points from X1~U(0,pi)
X1 = np.random.uniform(0,np.pi,ns1);
# transforming sampled points using f1
g1 = f1(X1);
print("Mean of samples =",round(np.mean(g1),2));
```

```python
ns2 = 50
# sampling ns2 points from X2~N(0,1/sqrt(2))
X2 = np.random.normal(0,2**(-0.5),ns2);
# taking absolute values of points
X2 = abs(X2);
# rejecting points beyond pi
X2 = X2[X2<np.pi];
# transforming sample points using f1
g2 = f1(X2);
print("Mean of samples =",round(np.mean(g2),2));
```
