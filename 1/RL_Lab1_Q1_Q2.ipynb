```python
import numpy as np
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
n1 = 100
a = np.random.multinomial(n1, [0.2, 0.4, 0.3, 0.1])
n2 = 1000
b = np.random.multinomial(n2, [0.2, 0.4, 0.3, 0.1])
n3 = 10000
c = np.random.multinomial(n3, [0.2, 0.4, 0.3, 0.1])
prob_a = a/n1
prob_b = b/n2
prob_c = c/n3
plt.stem(x, prob_a)
plt.title("Prob for n=100")
plt.xlabel("x")
plt.ylabel("Probabilities")
plt.show()
plt.stem(x, prob_b)
plt.title("Prob for n=1000")
plt.xlabel("x")
plt.ylabel("Probabilities")
plt.show()
plt.stem(x, prob_c)
plt.title("Prob for n=10000")
plt.xlabel("x")
plt.ylabel("Probabilities")
plt.show()
```

```python
n1 = 1000
no_of_samples = 10000
uniform_samples_mean = np.zeros(no_of_samples)
mean_uniform = 0.5
var_uniform = 1/12
for i in range(no_of_samples):
    a = np.random.uniform(0, 1, n1)
    uniform_samples_mean[i] = np.sum(a)/n1
Normalised_mean_distrib = (uniform_samples_mean - mean_uniform)*np.sqrt(no_of_samples)/(np.sqrt(var_uniform))
counts, bins = np.histogram(Normalised_mean_distrib)
plt.hist(bins[:-1], bins, weights=counts/no_of_samples, edgecolor="black") #This follows a standard normal distribution
plt.title("Standard_Normal from Uniform")
plt.xlabel("x")
plt.ylabel("Fraction of points")
plt.show()
```

```python
#Now solving for any mean and variance
mean = 5
variance = 4
Normal_Distribution = np.sqrt(variance)*Normalised_mean_distrib + mean
counts, bins = np.histogram(Normal_Distribution)
plt.hist(bins[:-1], bins, weights=counts/no_of_samples, edgecolor="black") #This follows a normal distribution
plt.title("Normal from the produced Standard_Normal")
plt.xlabel("x")
plt.ylabel("Fraction of points")
plt.show()
```
